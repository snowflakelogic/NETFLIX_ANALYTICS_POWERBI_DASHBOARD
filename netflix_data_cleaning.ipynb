```python
import pandas as pd
```

```python
netflix = pd.read_csv('netflix_titles.csv')
```

```python
netflix.head()
```

```python
netflix.isnull().sum()
```

```python
netflix.shape
```

```python
netflix['director']=netflix['director'].fillna('Unknown')
```

```python
netflix['cast']=netflix['cast'].fillna('Unknown')
```

```python
netflix['country'].unique
```

```python
netflix.isnull().sum()
```

```python
exploded = netflix.assign(country=netflix['country'].str.split(',')).explode('country')
```

```python
exploded.isnull().sum()
```

```python
exploded = netflix.assign(listed_in=netflix['listed_in'].str.split(',')).explode('listed_in')
```

```python
country_filled = exploded.groupby('listed_in')['country'].apply(
    lambda x: x.fillna(x.mode().iloc[0] if not x.mode().empty else 'Unknown')
).reset_index(drop=True)

exploded['country'] = country_filled
```

```python
exploded.head()
```

```python
exploded.isnull().sum()
```

```python
exploded['date_added'] = exploded.groupby('listed_in')['date_added'].apply(
    lambda x: x.fillna(x.mode().iloc[0] if not x.mode().empty else "Unknown")
).reset_index(drop=True)
```

```python
exploded.isnull().sum()
```

```python
exploded['rating'] = exploded.groupby('listed_in')['rating'].apply(
    lambda x: x.fillna(x.mode().iloc[0] if not x.mode().empty else "Unknown")
).reset_index(drop=True)
```

```python
exploded['duration'] = exploded.groupby('listed_in')['duration'].apply(
    lambda x: x.fillna(x.mode().iloc[0] if not x.mode().empty else "Unknown")
).reset_index(drop=True)
```

```python
exploded['country'] = exploded['country'].str.strip()
```
